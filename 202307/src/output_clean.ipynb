```python
import pandas as pd
```

```python
j = 1
df = pd.read_json(f'../outputs/generated_text_06_{j}.jsonl', lines=True, orient='records')
df_ = pd.read_json(f'../data/R{j}_short_answer_exam.json')
df
```

```python
df["complete"] = df["complete"].apply(lambda x: x[0])
df["answer"] = ""
for i in range(df.shape[0]):
    ans = df_[(df_["subject"] == df.loc[i, "subject"]) & (df_["Q_no"] == df.loc[i, "Q_no"])]["answer"].values[0]
    df.loc[i, "answer"] = ans
df
```

```python
df["correct"] = df["answer"] == df["complete"] 
print(df["correct"].value_counts())
df
```

```python
for j in range(1, 4):
    df = pd.read_json(f'../outputs/generated_text_06_{j}.jsonl', lines=True, orient='records')
    df_ = pd.read_json(f'../data/R{j}_short_answer_exam.json')
    df["complete"] = df["complete"].apply(lambda x: x[0])
    df["answer"] = ""
    for i in range(df.shape[0]):
        ans = df_[(df_["subject"] == df.loc[i, "subject"]) & (df_["Q_no"] == df.loc[i, "Q_no"])]["answer"].values[0]
        df.loc[i, "answer"] = ans
    df["correct"] = df["answer"] == df["complete"] 
    df.to_csv(f'../outputs/generated_text_06_{j}.csv', index=False, encoding='utf-8-sig')
    print(j)
    print(df["correct"].value_counts())
```
